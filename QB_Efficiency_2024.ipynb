```python
# Calculate average YDS per year
avg_yds_per_year = nfl.groupby('Year')['YDS'].mean()

# Plot
plt.figure(figsize=(10,6))
plt.plot(avg_yds_per_year.index, avg_yds_per_year.values, marker='o', label='Average Passing Yards')
plt.title('Average Passing Yards by Year')
plt.xlabel('Year')
plt.ylabel('Average YDS')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Calculate average TDs per year
avg_tds_per_year = nfl.groupby('Year')['TD'].mean()

# Plot with Year on X-axis
plt.figure(figsize=(10,6))
plt.plot(avg_tds_per_year.index, avg_tds_per_year.values, marker='o', label='Average Passing TDs')
plt.title('Average Passing TDs by Year')
plt.xlabel('Year')
plt.ylabel('Average TDs')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Sort by Year and YDS descending
top_qbs_per_year = nfl.sort_values(['Year', 'YDS'], ascending=[True, False])

# Get top 3 QBs for each year
top_3_qbs_by_year = top_qbs_per_year.groupby('Year').head(3)

# Display result
top_3_qbs_by_year[['Year', 'Name', 'Team', 'YDS']].sort_values(['Year', 'YDS'], ascending=[True, False])
```

```python
# Combine Year and Player for x-axis
top_3_qbs_by_year['Year_Player'] = top_3_qbs_by_year['Year'].astype(str) + ' - ' + top_3_qbs_by_year['Name']

# Plot
plt.figure(figsize=(14,8))
ax = sns.barplot(x='Year_Player', y='YDS', data=top_3_qbs_by_year, palette='Blues_d')

# Add data labels on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:,.0f}',                   # formatted with comma
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='bottom', 
                fontsize=10, color='black', xytext=(0, 3), 
                textcoords='offset points')

plt.title('Top 3 QBs by Passing Yards Per Year')
plt.xlabel('Year - Player')
plt.ylabel('Passing Yards')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter for 2024 QBs
nfl_2024 = nfl[nfl['Year'] == 2024].copy()

# Avoid division by zero
nfl_2024 = nfl_2024[nfl_2024['ATT'] > 0].copy()

# Compute passer rating components
a = ((nfl_2024['CMP'] / nfl_2024['ATT']) - 0.3) * 5
b = ((nfl_2024['YDS'] / nfl_2024['ATT']) - 3) * 0.25
c = (nfl_2024['TD'] / nfl_2024['ATT']) * 20
d = 2.375 - ((nfl_2024['INT'] / nfl_2024['ATT']) * 25)

# Clamp values between 0 and 2.375
a = a.clip(0, 2.375)
b = b.clip(0, 2.375)
c = c.clip(0, 2.375)
d = d.clip(0, 2.375)

# Calculate passer rating
nfl_2024['Efficiency'] = ((a + b + c + d) / 6) * 100
```

```python
top_10_qbs_rating = nfl_2024.sort_values('Efficiency', ascending=False).head(10)
top_10_qbs_rating[['Name', 'Team', 'CMP', 'ATT', 'YDS', 'TD', 'INT', 'Efficiency']]
```
