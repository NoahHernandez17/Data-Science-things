```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
passing = pd.read_html('https://www.nfl.com/stats/player-stats/')
passing[0]
```

```python
passing[0].describe().transpose()
```

```python
plt.figure(figsize=(10,6))
plt.plot(passing[0]['Player'],passing[0]['Pass Yds'], marker='o', label='Player Passing Yds')
plt.title('2025 Passing Yards / Player')
plt.xlabel('Player')
plt.ylabel('Total Yards')
plt.xticks(rotation = 90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(passing[0]['Player'],passing[0]['Yds/Att'], marker='o', label='Player Passing Yds')
plt.title('2025 Average Passing Yards / Player')
plt.xlabel('Player')
plt.ylabel('Total Yards')
plt.xticks(rotation = 90)
plt.yticks(rotation = 45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
```

```python
correlation = passing[0].corr(numeric_only = True)
plt.figure(figsize = (12,8))

sns.heatmap(correlation,
           annot = True,
           fmt = '.2f',
           cmap = 'coolwarm',
           linewidth = .5,
           linecolor = 'white',
           square = True)
plt.title('Correlation Heat map of NFL QB Stats')
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Get your data
x = passing[0]['Cmp']
y = passing[0]['Cmp %']

# Calculate r-value
r_value, _ = pearsonr(x, y)

# Plot with seaborn
sns.lmplot(x='Cmp', y='Cmp %', data=passing[0], height=5, aspect=1.5)
plt.title('Completions vs Completion %')

# Add r-value as text on the plot
plt.text(x=max(x)*0.6, y=min(y)*1.05, s=f'r = {r_value:.2f}', fontsize=12, color='darkred')

plt.tight_layout()
plt.show()
```
